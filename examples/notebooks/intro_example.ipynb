## Example introduction into a client with o6-python

In a notebook you have to await server responses, in a python script not

````python
from o6 import Client, StatusCodeError, types, Server
````

Set up a local server and a client on that server:

````python
s = Server(port=4840)
s.start()
c = Client("opc.tcp://localhost:4840")
await c.connect()
````

### Add a variable and change it

To handle variables or objects you always need the ```nodeid```.    
    - ```ns``` is always 1 for new variables and objects   
    - ```s``` is the string identifier or ```i``` as an integer identifier   

(no whitespaces in the nodeid string and divided with a semicolon)



````python
s.add_variable("Temperature", s.objects_node, 15.0, nodeid="ns=1;s=Temperature")
s.add_variable("Oven", s.objects_node, False, nodeid="ns=1;i=1001")

print(await c.read("ns=1;s=Temperature"))
````

Change the Temperature by using the nodeid. You have to use the o6-python types

````python
await c.write("ns=1;s=Temperature", types.Double(20))
print(await c.read("ns=1;s=Temperature"))
````

````python
await c.disconnect()
s.stop()
````
